# **Creating a dataframe from dictionary**

```python
weather_data = {
    'day':['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017'],
    'temperature':[32,35,28,24,32,31],
    'windspeed':[6,7,2,7,4,2],
    'event':['Rain','Sunny','Snow','Snow','Rain','Sunny']
}
```

```python
df2 = pd.DataFrame(weather_data)
df2
```

# **Creating a dataframe from list of tuples**

```python
weather_data = [
        ('1/1/2017',32,6,'Rain') ,
        ('1/2/2017',35,7,'Sunny'),
        ('1/6/2017',31,2,'Sunny')       

]
```

```python
df3 = pd.DataFrame(weather_data)
df3
```

# **changing the column of dataframe**

```python
df3 = pd.DataFrame(weather_data,columns=['day','temperature','windspeed','event'])
df3
```

# **Creating dataframe from list of dictionaries**

```python
weather_data = [
        {'day':'1/1/2017','temperature':32,'windspeed':6,'event':'Rain'} ,
        {'day':'1/2/2017','temperature':35,'windspeed':7,'event':'Sunny'},
        {'day':'1/6/2017','temperature':31,'windspeed':2,'event':'Sunny'}
]
```

```python
df4 = pd.DataFrame(weather_data)
df4
```
