# **Handle missing data using replace function**

```python
import pandas as pd
import numpy as np
```

```python
filename = 'https://raw.githubusercontent.com/codebasics/py/master/pandas/6_handling_missing_data_replace/weather_data.csv'

df = pd.read_csv(filename)
df
```

# **Replace value -99999 by NaN**

```python
new_df = df.replace(-99999,np.NaN)
new_df
```

# **Replace Multiple values by passing it as a list**

```python
new_df = df.replace([-99999,7],np.NaN)
new_df
```

# **Replace specific column with NaN values**

```python
df
#type(df['event'][3])
```

```python
new_df = df.replace({
    'temperature': -99999,
    'windspeed': -99999,
    'event': '0'
},np.NaN)

new_df
```

# **Replace -99999 in temperature and windspeed column by Nan and 0 in event column with sunny**

```python
df
```

```python
new_df = df.replace({
  -99999 : np.NaN,
        '0':'Sunny'
})
new_df
```

```python
df
```

# **Replace Units entry such as F and MPH from the dataframe**

```python
new_df = df.replace({
  32 : '32 F',
  6 : '6 MPH'
})
new_df
```

```python
new_df
```

```python
new_df_1 = new_df.replace('[A-Za-z]','',regex=True)
new_df_1
```

# **It replaces eevent string with also blank space so we can pass only the column by passing it as a dictionary**

```python
new_df = new_df.replace({
    -99999:np.NaN,
    '0':'Sunny'
})
new_df
```

```python
new_df = new_df.replace({
    'temperature': '[A-Za-z]',
    'windspeed':  '[A-Za-z]'
},'',regex=True)
new_df
```
