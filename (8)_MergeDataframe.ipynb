# **Merge Dataframes**

```python
import pandas as pd
```

```python
Germany_weather = pd.DataFrame({
    "city":["frankfurt","berlin","munich"],
    "temperature":[15,18,13],
    "humidity":[14,15,16]
})
Germany_weather
```

```python
Germany_weather_w = pd.DataFrame({
    "city":["munich","frankfurt","berlin"],
    "windspeed":[150,180,130],
})
Germany_weather_w
```

```python
new_df = pd.merge(Germany_weather,Germany_weather_w,on="city")
new_df
```

```python
weather_event = pd.DataFrame({
    "city":["berlin","munich","frankfurt"],
    "event":["dry","humid","rainy"]
})
weather_event
```

```python
new_df = pd.merge(Germany_weather,Germany_weather_w,on="city")
new_df
```

# **joins in dataframes**

```python
Germany_weather = pd.DataFrame({
    "city":["frankfurt","berlin","munich","stutgart","cologne","erlangen"],
    "temperature":[15,18,13,12,17,19],
    "humidity":[14,15,16,22,24,25]
})
Germany_weather
```

```python
weather_event = pd.DataFrame({
    "city":["berlin","munich","frankfurt","hamburg"],
    "event":["dry","humid","rainy","fog"]
})
weather_event
```

```python
new_df = pd.merge(Germany_weather,weather_event,on="city")
new_df
```

```python
new_df = pd.merge(Germany_weather,weather_event,on="city",how="inner")
new_df
```

```python
new_df = pd.merge(Germany_weather,weather_event,on="city",how="outer")
new_df
```

```python
new_df = pd.merge(Germany_weather,weather_event,on="city",how="left")
new_df
```

```python
new_df = pd.merge(Germany_weather,weather_event,on="city",how="right")
new_df
```

# **you know from where data came in joins**

```python
new_df = pd.merge(Germany_weather,weather_event,on="city",how="outer",indicator=True)
new_df
```
