# **Handle Missing Data in pandas by such as fillna,drop,interpolate**

```python
import pandas as pd
```

```python
filename = "https://raw.githubusercontent.com/codebasics/py/master/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv"
```

```python
df = pd.read_csv(filename)
df
```

```python
type(df.day[0])
```

# **change day datatype from string to timestamp using parse_dates**

```python
df = pd.read_csv(filename,parse_dates=['day'])
df.set_index('day',inplace=True)
df
```

# **fill Nan values with 0**

```python
new_df = df.fillna(0)
new_df
```

# **fill na values of specific column with any specific values**


```python
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

# **copy previous row value to current zero value**

```python
new_df = df.fillna(method='ffill')
new_df
```

# **copy next row value to current row value**

```python
df
```

```python
new_df = df.fillna(method='bfill')
new_df
```

# **copy next column value to current column value**

```python
df
```

```python
new_df = df.fillna(method='bfill',axis='columns')
new_df
```

# **copy previous row value to current row with limit that how many times it should copy**

```python
df
```

```python
new_df = df.fillna(method='ffill',limit=1)
new_df
```

```python
new_df = df.fillna(method='ffill',limit=3)
new_df
```

# **Instead of filling na value with guess we do interpolation and also with time method**

```python
df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method='time')
new_df
```

# **drop na values**

```python
df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how="all")
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```
