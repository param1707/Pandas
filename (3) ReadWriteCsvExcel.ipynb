# **Reading a csv file and converting all the non available values to Nan**

```python
df = pd.read_csv('https://raw.githubusercontent.com/param1707/Pandas/main/stock_data_n.csv',header=None,names=['tickers','eps','revenue','price','people'])
df
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/param1707/Pandas/main/stock_data_n.csv',header=None,names=['tickers','eps','revenue','price','people'],na_values=["not available","n.a."])
df
```

# **Converting the specific values of column to NaN instead of passing as a list we pass it as by passing na_values parameters as a dictionary**

```python
df = pd.read_csv('https://raw.githubusercontent.com/param1707/Pandas/main/stock_data.csv',na_values={
    'eps':["not available","n.a."],
    'revenue':["not available","n.a.",-1],
    'price':["not available","n.a."],
    'people':["not available","n.a."]
})
df
```

# **Reading a excel file with sheet name**

```python
df = pd.read_excel("stock_data.xlsx",sheet_name="Sheet1")
df
```

# **changing na and not available value using converter function**

```python

def convert_people_cell(cell):
  if cell == 'n.a.':
    return "sam walton"
  return cell

def convert_eps_cell(cell):
  if cell == 'not available':
    return None
  return cell


df = pd.read_excel("stock_data.xlsx",sheet_name="Sheet1",converters={
    'people': convert_people_cell,
    'eps': convert_eps_cell
})
df
```

# **Writing a dataframe to an excel file**

```python
df.to_excel("stock_data_new.xlsx",sheet_name="stocks",index=False)
```

```python
df2 = pd.read_excel("stock_data_new.xlsx",sheet_name="stocks")
df2
```

# **writing dataframe to an excel file using an offset**




```python
df.to_excel("stock_data_new_offset.xlsx",sheet_name="stocks",index=False,startrow=2,startcol=2)
```

```python
df2 = pd.read_excel("stock_data_new_offset.xlsx",sheet_name="stocks")
df2
```
