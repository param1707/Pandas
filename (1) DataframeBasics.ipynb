# **Read a datafarme from a dictionary such that key will become columns and value will become rows of dataframe**

```python
weather_data = {
    'day':['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017'],
    'temperature':[32,35,28,24,32,31],
    'windspeed':[6,7,2,7,4,2],
    'event':['Rain','Sunny','Snow','Snow','Rain','Sunny']
}
```

```python
weather_data
```

```python
df = pd.DataFrame(weather_data)
df
```

```python
df.shape
```

```python
rows,column = df.shape
print("rows:",rows)
print("columns:",column)
```

# **printing the particular row of dataframes using some sort of condition like sql**

```python
df[['day','temperature','windspeed','event']]
```

```python
df[['day','temperature','windspeed','event']][df.temperature > 30]
```

```python
df[['day','temperature','windspeed','event']][df.temperature == df.temperature.max()]
```

```python
df[['day','temperature','windspeed','event']][df.day == '1/6/2017']
```

```python
df[['day','temperature','windspeed','event']][df['windspeed'] == 7]
```

```python
df[['day','temperature','windspeed','event']][df['event'] == 'Rain']
```
