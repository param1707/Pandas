# **concate of two or more pandas dataframes**

```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
    "city":["mumbai","delhi","calcutta"],
    "temperature":[45,40,35],
    "humidity":[40,50,60]
})
india_weather
```

```python
Germany_weather = pd.DataFrame({
    "city":["frankfurt","berlin","munich"],
    "temperature":[15,18,13],
    "humidity":[14,15,16]
})
Germany_weather
```

```python
new_df = pd.concat([india_weather,Germany_weather])
new_df
```

# **Inorder to change index**

```python
new_df = pd.concat([india_weather,Germany_weather],ignore_index=True)
new_df
```

```python
new_df = pd.concat([india_weather,Germany_weather],ignore_index=True,keys=["INDIA","GERMANY"])
new_df
```

# **Remove ignore_index**

```python
new_df = pd.concat([india_weather,Germany_weather],keys=["INDIA","GERMANY"])
new_df
```

# **using keys we can access sub dataframe from a main dataframe**

```python
new_df.loc["INDIA"]
```

```python
new_df.loc["GERMANY"]
```

# **concate two data frame along the column**

```python
Germany_weather = pd.DataFrame({
    "city":["frankfurt","berlin","munich"],
    "temperature":[15,18,13],
    "humidity":[14,15,16]
})
Germany_weather
```

```python
Germany_weather_w = pd.DataFrame({
    "city":["frankfurt","berlin","munich"],
    "windspeed":[150,180,130],
})
Germany_weather_w
```

```python
new_df = pd.concat([Germany_weather,Germany_weather_w])
new_df
```

```python
new_df = pd.concat([Germany_weather,Germany_weather_w],axis=1)
new_df
```

# **suppose my data is in different order**

```python
Germany_weather_w = pd.DataFrame({
    "city":["munich","frankfurt","berlin"],
    "windspeed":[150,180,130],
})
Germany_weather_w
```

```python
Germany_weather = pd.DataFrame({
    "city":["frankfurt","berlin","munich"],
    "temperature":[15,18,13],
    "humidity":[14,15,16]
})
Germany_weather
```

```python
new_df = pd.concat([Germany_weather,Germany_weather_w],axis=1)
new_df
```

```python
Germany_weather_w = pd.DataFrame({
    "city":["munich","frankfurt","berlin"],
    "windspeed":[150,180,130],
},index=[2,0,1])
Germany_weather_w
```

```python
Germany_weather = pd.DataFrame({
    "city":["frankfurt","berlin","munich"],
    "temperature":[15,18,13],
    "humidity":[14,15,16]
},index=[0,1,2])
Germany_weather
```

```python
new_df = pd.concat([Germany_weather,Germany_weather_w],axis=1)
new_df
```

# **concating a series and dataframe**

```python
new_s = pd.Series(["humid","dry","rainy"],name="event")
new_s
```

```python
Germany_weather_w = pd.DataFrame({
    "city":["munich","frankfurt","berlin"],
    "windspeed":[150,180,130],
},index=[2,0,1])
Germany_weather_w
```

```python
new_df = pd.concat([Germany_weather_w,new_s],axis=1)
new_df
```
